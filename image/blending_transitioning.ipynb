```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
%matplotlib notebook
```

```python
image_path = "data//image//lena_color_512.tif"
image_1=cv2.imread(image_path)
image_1=cv2.cvtColor(image_1,cv2.COLOR_BGR2RGB)
```

```python
image_path = "data//image//peppers_color.tif"
image_2=cv2.imread(image_path)
image_2=cv2.cvtColor(image_2,cv2.COLOR_BGR2RGB)
```

```python
def blending(image_1,alpha,image_2,beta,gamma):
    return ((image_1*alpha).astype(np.uint8)+(image_2*beta).astype(np.uint8)+gamma)
```

```python
alpha=0.3
beta=1-alpha
gamma=0
```

```python
plt.imshow(blending(image_1,alpha,image_2,beta,gamma))
plt.show()
```

```python
fig = plt.figure()
ims = []
for i in np.linspace(0,1,10):
    alpha=i
    beta=1-alpha
    gamma=0
    im = plt.imshow(blending(image_1,alpha,image_2,beta,gamma), animated=True)
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval=500, blit=True, repeat_delay=1000)
plt.show()
```
